```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go
```

```python
# Define toroidal coordinates
theta = np.linspace(0, 2*np.pi, 100)
phi = np.linspace(0, 2*np.pi, 100)
theta, phi = np.meshgrid(theta, phi)
R = 1  # Major radius
r = 0.3  # Minor radius
```

```python
# Convert toroidal coordinates to Cartesian coordinates
x = (R + r * np.cos(theta)) * np.cos(phi)
y = (R + r * np.cos(theta)) * np.sin(phi)
z = r * np.sin(theta)
```

```python
%matplotlib notebook
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the torus
ax.plot_surface(x, y, z, cmap='viridis', alpha=0.8)

# Plot the toroidal axes
toroidal_axis_length = 1.5 * (R + r)
ax.plot([-toroidal_axis_length, toroidal_axis_length], [0, 0], [0, 0], 'k-', linewidth=2)
ax.plot([0, 0], [-toroidal_axis_length, toroidal_axis_length], [0, 0], 'k-', linewidth=2)
ax.plot([0, 0], [0, 0], [-r - toroidal_axis_length, -r + toroidal_axis_length], 'k-', linewidth=2)

# Set plot parameters
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Toroidal Coordinates')

plt.show()
```

```python
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, 
                                 colorscale="plasma_r", showscale=False)])

fig.show()
```
