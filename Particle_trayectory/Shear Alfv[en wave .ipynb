```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,50)
d = [-2.5,-1.25,0,1.25,2.5]
```

```python
def sen_func(x,d):
    return np.sin(x) + d
```

```python
f0 = sen_func(x,0)
```

```python
%matplotlib notebook
fig = plt.figure(figsize=(9,4))

for i in d:
    plt.axhline(i, color="black",linestyle = "--",alpha = 0.7)
    
plt.axhline(0,color="red",linestyle = "-")    
for i in d:
    plt.plot(x,sen_func(x,i),color = "dodgerblue")
    
    
    
plt.annotate(r"""$\textbf{B}_0$""",xy = (3.5,2.7))
#plt.annotate(r"""$\tilde{B}$""", xy = (3,-3.5), color = "dodgerblue")
plt.annotate(r"""$\textbf{k}$""", xy = (-0.2,0.05), color = "red")
plt.title(f"Shear AlfvÃ©n Waves") 
plt.ylabel(r"""Magnetic Field ($\textbf{B}$)""")
plt.show()
```
