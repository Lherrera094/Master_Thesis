Imports library booz_xform (Source: https://pypi.org/project/booz-xform/)

```python
import booz_xform as bx
import matplotlib.pyplot as plt
import os
```

Setups saving files

```python
device = "TJ-II"
profile = "Exp"
save_file = f"{device}_{profile}_boozer"
if not os.path.exists(save_file):
    os.mkdir(save_file)
```

Creates a Booz_xform instance

```python
b = bx.Booz_xform()
```

Reads wout file

```python
b.read_wout(f"wout_files/wout_tj2curref.nc")
```

$\iota$ profile

```python
abs(b.iota)
```

Selects the toroidal ($\textbf{n}$) and poloidal ($\textbf{m}$) modes for transformation

```python
#b.mboz = 12
#b.nboz = 18
```

Transform to Boozer Coordinates

```python
b.compute_surfs
b.run()
```

Plot Magnetic field on specified surface ($js$)

```python
%matplotlib notebook
figure = bx.surfplot(b,js=95,cmap=plt.cm.jet)

plt.title(r"$|B|$ on s = " + f"0.955")
plt.xlabel(r"Toroidal Angle $\varphi_B$")
plt.ylabel(r"Poloidal Angle $\theta_B$")
plt.savefig(f"{save_file}/Heatmap_B.png")
plt.show(figure)
```

```python
%matplotlib notebook
figure = bx.modeplot(b, legend_args={"fontsize":4})
plt.show(figure)
```

Creates file in to boozer coordinates

```python
b.write_boozmn(f"{save_file}/boozmn_{device}_{profile}.nc")
```
